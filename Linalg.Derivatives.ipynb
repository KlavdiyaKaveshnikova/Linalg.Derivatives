## Уровень 0

посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
import numpy as np
```

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

$f'(x) = 5x^4 + 8cos(2x) - 3sin(3x+3)$

$f'(1) = 5 + 8cos(2) - 3sin(6) = 5 + 8* (-0,4161) - 3*(-0,2794) = 5 - 3,3288 + 0,8382 = 2,509$

$f''(x) = 20x^3 - 16sin(2x) - 9cos(3x+3)$

$f''(1) = 20 - 16sin(2) - 9cos(6) = 20 -16*(0,9093) - 9*(0,9602) = 20 - 14,5488 - 8,6418 = -3,1906$

```python
from scipy.misc import derivative
def f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x+3)

print(derivative(f, 1.0, dx=1e-6))
print(derivative(f, 1.0, dx=1e-6, n=2))
```

## Уровень 1

Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]

```python
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (np.sin(2*x+1))**5
```

```python
x = np.linspace(-5, 5)
```

```python
fx = f(x)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
f2x = [derivative(f, xx, dx=1e-6, n=2) for xx in x]
```

```python
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='first derivative')
plt.plot(x, f2x, label='second derivative')
plt.legend()
plt.show()
```
